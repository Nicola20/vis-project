```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("data/vgsalesglobale.csv")
data
```

# Genres

```python
genreCount = [len(data.loc[data["genre"] == genre]) for genre in data.genre.unique()]
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = genreCount, y = data.genre.unique(), orient = "h")
```

```python
publisherCount = [len(data.loc[data["publisher"] == publisher]) for publisher in data.publisher.unique()]
plt.figure(figsize = (15, 100))
sns.barplot(x = publisherCount, y = data.publisher.unique(), orient = "h")
```
